```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from numpy import ndarray
import math
%matplotlib inline

from typing import Callable
from typing import Dict
from typing import List
```

```python
# define ndarray function chain
Array_Function=Callable[[ndarray],ndarray]
Chain=List[Array_Function]

#define calling multi-chained functions
def chain_call(chain:Chain,input_:ndarray)->ndarray:
    output_=input_  
    for func in chain:
        output_=func(output_)
    return output_



#define derivative of single or chain of Array_Functions
def deriv(func:Array_Function,input_:ndarray,delta:float=0.001)->ndarray:
    return (func(input_+delta)-func(input_-delta))/(2*delta)

#define 2_level chained deriv
 #   given f2(f1(x)), a two level chained function,find derivative:  
  #  f2'(f1(input_))*f1'(input_)
def chain_deriv_2(chain:Chain,input_:ndarray)->ndarray:
    #compute f1'(input_)
    f1dx=deriv(chain[0],input_);
    
    #compute f2'(f1(input_))
    f1=chain[0](input_)
    f2du=deriv(chain[1],f1)
    
    return f2du*f1dx


```

```python
#test testing chain call
def f1(x:ndarray)->ndarray:
    return 2+x
def f2(x:ndarray)->ndarray:
    return x*x

chain=[f1,f2]
x=np.arange(-3,3,1.0)
y=chain_call(chain,x)


```
