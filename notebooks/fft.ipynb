```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

Fs = 2048;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

ff = 20;   # frequency of the signal
y1=np.sin(np.pi*ff*t)
y2=np.sin(2*np.pi*ff*t)
y3=np.sin(3.5*np.pi*ff*t)
y = y1+y2+y3

n = len(y) # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(int(n/2))] # one side frequency range

Y = np.fft.fft(y)/n # fft computing and normalization
Y = Y[range(int(n/2))]

fig, ax = plt.subplots(5, 1)
fig.set_size_inches(10,28)


ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude y(x)')

frq1=frq[range(100)]
Y1 = Y[range(int(100))]
ax[1].plot(frq1,abs(Y1),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')

ax[2].plot(frq,Y.real,'r') # plotting the spectrum
ax[2].set_xlabel('Freq (Hz)')
ax[2].set_ylabel('Y.real')

ax[3].plot(frq,Y.imag,'r') # plotting the spectrum
ax[3].set_xlabel('Freq (Hz)')
ax[3].set_ylabel('Y.imag')

ph=np.arctan(Y.imag/Y.real)
ax[4].plot(frq,ph,'r') # plotting the spectrum
ax[4].set_xlabel('Freq (Hz)')
ax[4].set_ylabel('Phase')


fig.show()                                     
```
